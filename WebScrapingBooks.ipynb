```python
import requests
import lxml
import bs4
```

```python
result = requests.get("https://books.toscrape.com/index.html")
result
```

```python
soup = bs4.BeautifulSoup(result.text,"lxml")
soup
```

```python
#Lista todas as url que contém as categorias de books para acessar todas as páginas
lhref = []
for a in soup.find_all('a', href=True):
    print("Found the URL:", a['href'])
    lhref.append(a['href'])
lhref
```

```python
#Seleciona apenas as páginas das categorias
base_url = []
for i in lhref:
    if 'catalogue/category/books/' in i:
        base_url_aux = 'https://books.toscrape.com/' + i
        base_url.append(base_url_aux)
base_url
```

```python
# "([a-zA-Z]+(-[a-zA-Z]+)*)_"g
```

```python
from datetime import datetime
start_time = datetime.now()
start_time
```

```python
dict_1 = {}

def tableBooks(soup_1, dict_cat):
    if not 'title' in dict_cat:
        dict_cat['title'] = []
    if not 'rate' in dict_cat:     
        dict_cat['rate'] = []
    for j in range(0, len(soup_1.select(".product_pod"))):
        title_aux = soup_1.select(".product_pod")[j].select("a")[1]["title"]
        rating_aux = soup_1.select(".product_pod")[j].select(".star-rating")[0]['class'][1]
        dict_cat['title'].append(title_aux)
        dict_cat['rate'].append(rating_aux)    
  
for i in range(0,len(base_url)):
    req_1 = requests.get(base_url[i])
    soup_1 = bs4.BeautifulSoup(req_1.text,"lxml")
    category = base_url[i].split('/')[-2].split('_')[0]  
    print('Category:', category)
    
    #First page
    dict_1[category] = {}
    tableBooks(soup_1, dict_1[category])

    #Next pages
    while not [] == soup_1.select(".pager > .next > a"):
        print("Página:", soup_1.select(".pager > .next > a")[0]['href'])
        req_1 = requests.get(base_url[i][0:-10] + soup_1.select(".pager > .next > a")[0]['href'])
        soup_1 = bs4.BeautifulSoup(req_1.text,"lxml")
        tableBooks(soup_1, dict_1[category]) 
```

```python
end_time = datetime.now()
end_time
```

```python
import json
# Convert the dictionary to a JSON string with double quotes
json_string = json.dumps(dict_1, indent=4)  
print(json_string)
```

```python
# the json file to save the output data   
save_file = open("ScrapBook.json", "w")  
json.dump(json_string, save_file)  
save_file.close() 
```
